```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python

# getting information about the site
request = requests.get("https://realpython.github.io/fake-jobs/")

# creating the soup
soup = BeautifulSoup(request.content, "html.parser")
result = soup.find_all("div", class_="card-content")
```

```python
# creating lists for temporary storage
positions = list()
companies = list()
addresses = list()
dates = list()

# putting the information we need on the right list
for i in result:
    title = i.find("h2", class_="title").text.strip()
    positions.append(title)
    brand = i.find("h3", class_="company").text.strip()
    companies.append(brand)
    location = i.find("p", class_="location").text.strip()
    addresses.append(location)
    date = i.find("p", class_="is-small").text.strip()
    dates.append(date)
```

```python
# creating a dataframe to display the results in a table
df = pd.DataFrame({
    "Job Title": positions,
    "Company name": companies,
    "Business address": addresses,
    "Publications Date": dates,
})

df.head()
```
